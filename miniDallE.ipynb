```python
import googletrans
from googletrans import Translator
from min_dalle import MinDalle
```

**Resmin Açıklaması**

Bu kısımda oluşturulmak istenen resmin açıklaması yazılır.

```python
yazi = "Ayçiçeği tarlasında ekim yapan işçi, yağlı boya"
```

```python
translator = Translator()
dil = translator.detect(yazi).lang
ingilizceye_cevrilmis = translator.translate(yazi, src=dil, dest= 'en')
ceviri= ingilizceye_cevrilmis.text
print(ceviri)
```

```python
model = MinDalle(is_mega=True, is_reusable=True)
```

**Resmin Oluşturulması**

```python
yazi = ceviri
cekirdek_seed = 4
grafik_boyutu = 1

display(model.generate_image(yazi, cekirdek_seed, grafik_boyutu))
```
